```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
img_gray=cv2.imread("./cameraman.png",0)
histr=cv2.calcHist([img_gray],[0],None,[256],[0,256])
plt.plot(histr)
plt.show()
```

```python
img_gray=cv2.imread("./cameraman.png",0)
img_eq_hist=cv2.equalizeHist(img_gray)

fig=plt.figure(figsize=(12,5))
rows = 1
columns = 2

fig.add_subplot(rows,columns,1)
plt.imshow(img_gray,cmap="gray")
plt.axis("off")
plt.title("Orginal Image")

fig.add_subplot(rows,columns,2)
plt.imshow(img_eq_hist,cmap="gray")
plt.axis("off")
plt.title("BINARY THERESHOLDING Image")
```

```python
hist_org=cv2.calcHist([img_gray],[0],None,[256],[0,256])
hist_eq=cv2.calcHist([img_eq_hist],[0],None,[256],[0,256])

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(hist_org,color="b")
plt.title("Orjinal Grnt Histogram")


plt.subplot(1,2,2)
plt.plot(hist_eq,color="b")
plt.title("Histogram Eitleme sonucu  Histogram")
```

```python
Clache=cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
mg_adaptive_eq_hist=Clache.apply(img_gray)
fig=plt.figure(figsize=(15,5))
rows = 1
columns = 3

fig.add_subplot(rows,columns,1)
plt.imshow(img_gray,cmap="gray")
plt.axis("off")
plt.title("Orginal Image")

fig.add_subplot(rows,columns,2)
plt.imshow(img_eq_hist,cmap="gray")
plt.axis("off")
plt.title("BINARY THERESHOLDING Image")

fig.add_subplot(rows,columns,3)
plt.imshow(mg_adaptive_eq_hist,cmap="gray")
plt.axis("off")
plt.title("Adaptive Image")

```
